```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pale_yellow = sns.light_palette('#FAE67C', n_colors=4, reverse=True)
sns.palplot(pale_yellow)
```

```python


data = {
    'funnel': ['Marketing', 'Trial', 'Sales', 'Paying'],
    'dropoff': [100, 87, 87/2 - 10, 25]
}
plt.figure(figsize=(9, 6))

# df = pd.DataFrame(pd.Series(data=data)).reset_index()
# df.rename(columns={'index': 'step', 0: 'dropoff'}, inplace=True)
df=pd.DataFrame.from_dict(data)

# df

sns.set_style('whitegrid')
sns.set(font="Consolas", context='talk', style='whitegrid')
sns.set_palette(pale_yellow)
ax = sns.barplot(data=df, x='funnel', y='dropoff')
# plt.set(xlabel='Sales Funnel Step')
ax.set(xlabel=None, ylabel='Drop-off (%)')
sns.despine()
plt.title('Drop-off by Sales Funnel Step', fontsize=28)
plt.savefig('funnel-dropoff.svg', format='svg')
```

```python
fig = plt.figure(figsize=(16,6))
ax1 = fig.add_subplot(222)

data1 = {
    'option': ['Both', 'Not sure', 'Read docs', 'Manage docs'],
    'number_users': [960, 310, 260, 245]
}

df1 = pd.DataFrame.from_dict(data1)
sns.barplot(data=df1, x='option', y='number_users', ax=ax1)
sns.despine()
ax1.title.set_text('Reality')
ax1.set(ylabel=None, xlabel="Option")

ax2 = fig.add_subplot(221)

data2 = {
    'option': ['Both', 'Not sure', 'Read docs', 'Manage docs'],
    'number_users': [100, 100, 600, 1000]
}
df2 = pd.DataFrame.from_dict(data2)
sns.barplot(data=df2, x='option', y='number_users', ax=ax2)
ax2.title.set_text('Expectations')
ax2.set(ylabel="Number of Users", xlabel="Option")

plt.savefig('expectations-v-reality.svg', format='svg')
```
