```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# sns.set_style(rc={'font.family': 'Times New Roman'})
# sns.set_context('talk')
sns.set(font="Liberation Sans", context='talk')
sns.axes_style()
```

```python
import matplotlib as mpl
font_paths = mpl.font_manager.findSystemFonts()
font_objects = mpl.font_manager.createFontList(font_paths)
font_names = [f.name for f in font_objects]
# print(font_names)
```

```python
data = {
    'funnel': ['Marketing', 'Trial', 'Sales', 'Paying'],
    'dropoff': [100, 87, 87/2 - 10, 25]
}
plt.figure(figsize=(9, 6))

# df = pd.DataFrame(pd.Series(data=data)).reset_index()
# df.rename(columns={'index': 'step', 0: 'dropoff'}, inplace=True)
df=pd.DataFrame.from_dict(data)

# df

sns.set_style('whitegrid')
sns.set(font="Consolas", context='talk', style='whitegrid')
blues = sns.color_palette("Blues")
sns.set_palette(blues)
ax = sns.barplot(data=df, x='funnel', y='dropoff')
# plt.set(xlabel='Sales Funnel Step')
ax.set(xlabel='Sales Funnel Step', ylabel='Drop-off (%)')
sns.despine()
plt.title('Drop-off by Sales Funnel Step', fontsize=28)
plt.savefig('bargraph.svg', format='svg')
# # sns.
```
